#### EMS and ZIP Codes
Review of April through August averaged across 5 years. 

```python
nz = ems[ems['MONTH'].isin(list(range(4,8,1)))]
nz = nz.groupby(['ZIPCODE']).agg(
    {'calls':'sum', 'its':[np.mean], 'irs':[np.mean], 'TEMP':[np.mean]})
nz.reset_index(inplace=True)
nz.columns = ['_'.join(col).strip() for col in nz.columns.values]
nz.sort_values('its_mean', inplace = True)
nz = nz.assign(ZIPCODE= nz.ZIPCODE_.astype(str))
nz = nz.assign(LITS= np.log(nz.its_mean))

nz.info()
```

#### Preview of the Primary GeoJSON Data

The below plot demonstrates that the converted `.shp` file correctly displays from the `.geojson` url.  

```python
j = "https://raw.githubusercontent.com/jbrnbrg/DATA608Submissions/master/data/nyczipcodes.geojson"
import requests
import fiona
import geopandas as gpd

request = requests.get(j)
b = bytes(request.content)
with fiona.BytesCollection(b) as f:
    crs = f.crs
    gdf = gpd.GeoDataFrame.from_features(f, crs=crs)

f, ax = plt.subplots(1, figsize=(12, 10))

gdf = gdf.to_crs({'init': 'epsg:3395'}) # mercator.
gdf = gdf.merge(nz, how = 'left', on = 'ZIPCODE')
plt.rcParams["xtick.labelsize"] = 10
plt.rcParams["ytick.labelsize"] = 10
plt.title('Incident Travel Seconds (months April to August, averaged across 5 years)')
gdf.plot(column='its_mean', ax=ax, legend = True)
```

```python
ff, axx = plt.subplots(1, figsize=(12, 10))
plt.title('Incident Response Seconds (months April to August, averaged across 5 years)')
gdf.plot(column='irs_mean', ax=axx, legend = True)
```
